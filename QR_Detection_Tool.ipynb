```python
import cv2
import time
from threading import Thread, Lock
from queue import Queue
from pyzbar.pyzbar import decode
```

```python
class QRScanner:
    def __init__(self):
        self.cam = cv2.VideoCapture(0)
        self.cam.set(3, 640)  # Set width
        self.cam.set(4, 480)  # Set height
        #We setted the witdh & height to 640x480 pixels
        self.qr_data_queue = Queue()
        self.qr_data_lock = Lock()
        self.scan_status = "Scanning..."
        
#This functoins initaites the scanning process
    def start_scanning(self):
        Thread(target=self.capture_frames).start()
        Thread(target=self.process_qr_codes).start()
        self.display_ui()

#This function keeps capturing frames from the webcam in a continuous loop.
    def capture_frames(self):
        while True:
            success, frame = self.cam.read()
            if not success:
                print("Error: Failed to capture frame.")
                break
            self.qr_data_lock.acquire()
            self.qr_data_queue.put(frame)
            self.qr_data_lock.release()

#This function continuously processes frames containing QR codes from the queue.
    def process_qr_codes(self):
        while True:
            if not self.qr_data_queue.empty():
                self.qr_data_lock.acquire()
                frame = self.qr_data_queue.get()
                self.qr_data_lock.release()

                decoded_objects = decode(frame)
                for obj in decoded_objects:
                    self.process_detected_qr_code(obj)

#This function processes a single detected QR code. It decodes the data encoded in the QR code using the data attribute of the qr_code object and decodes it using UTF-8 encoding.

    def process_detected_qr_code(self, qr_code):
        qr_data = qr_code.data.decode('utf-8')
        print("Type:", qr_code.type)
        print("Data:", qr_data)

        

    def display_ui(self):
        while True:
            self.qr_data_lock.acquire()
            if not self.qr_data_queue.empty():
                frame = self.qr_data_queue.queue[-1]
            else:
                frame = None
            self.qr_data_lock.release()

            if frame is not None:
                self.display_frame_with_qr_data(frame)
            else:
                print("Error: No frame available.")

            key = cv2.waitKey(4)
            # Press q to stop/break the loop
            if key == ord('q'):
                break

        self.cam.release()
        cv2.destroyAllWindows()

    def display_frame_with_qr_data(self, frame):
        cv2.putText(frame, self.scan_status, (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)
        cv2.imshow("OurQR_code_scanner", frame)

if __name__ == "__main__":
    scanner = QRScanner()
    scanner.start_scanning()
```
